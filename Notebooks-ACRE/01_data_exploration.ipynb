## 4. Missing Values Analysis

```python
# Calculate missing percentages
missing_data = pd.DataFrame({
    'Column': df_raw.columns,
    'Missing_Count': df_raw.isnull().sum(),
    'Missing_Percentage': (df_raw.isnull().sum() / len(df_raw)) * 100
}).sort_values('Missing_Percentage', ascending=False).reset_index(drop=True)

print("Missing Values Analysis:\n")
print(missing_data[missing_data['Missing_Percentage'] > 0])

# Check for zero values in budget/revenue (likely missing)
budget_zeros = (df_raw['budget'] == 0).sum()
revenue_zeros = (df_raw['revenue'] == 0).sum()

print(f"\nBudget = 0: {budget_zeros:,} ({100*budget_zeros/len(df_raw):.2f}%)")
print(f"Revenue = 0: {revenue_zeros:,} ({100*revenue_zeros/len(df_raw):.2f}%)")
print("\n⚠ These zeros likely represent missing data, not actual zero budgets/revenues")
```

```python
# Visualize missing data (SAVE)
missing_plot = missing_data[missing_data['Missing_Percentage'] > 0].head(10)

fig = go.Figure([
    go.Bar(
        x=missing_plot['Column'],
        y=missing_plot['Missing_Percentage'],
        marker_color=GOLD,
        marker_line_color=BLACK,
        marker_line_width=1
    )
])

fig.update_layout(
    template=plotly_template,
    title='Top 10 Columns with Missing Data',
    xaxis_title='Column',
    yaxis_title='Missing Percentage (%)',
    xaxis_tickangle=-45,
    height=500,
    showlegend=False
)

fig.write_html(CHART_DIR / '01_missing_data.html')
fig.show()
print(f"✓ Saved: {CHART_DIR / '01_missing_data.html'}")
```

## 5. Categorical Features Analysis

### 5.1 Status & Language Distribution

```python
# Status value counts
status_counts = df_raw['status'].value_counts()

print("Movie Status Distribution:\n")
print(status_counts)
print(f"\n✓ {100*status_counts['Released']/len(df_raw):.2f}% of movies are 'Released'")

# Adult content distribution
adult_counts = df_raw['adult'].value_counts()
print("\nAdult Content Flag Distribution:\n")
print(adult_counts)
print(f"\n✓ {100*adult_counts.get(False, 0)/len(df_raw):.2f}% are non-adult content")
```

```python
# Top 15 languages (SAVE)
top_languages = df_raw['original_language'].value_counts().head(15)

fig = go.Figure([
    go.Bar(
        y=top_languages.index,
        x=top_languages.values,
        orientation='h',
        marker_color=GOLD,
        marker_line_color=BLACK,
        marker_line_width=1
    )
])

fig.update_layout(
    template=plotly_template,
    title='Top 15 Original Languages',
    xaxis_title='Number of Movies',
    yaxis_title='Language',
    height=600,
    showlegend=False
)

fig.write_html(CHART_DIR / '02_top_languages.html')
fig.show()
print(f"✓ Saved: {CHART_DIR / '02_top_languages.html'}")
```

### 5.2 Genre Analysis

```python
# Parse genres and count frequencies
all_genres = []

for genres_str in df_raw['genres'].dropna():
    if isinstance(genres_str, str):
        genres = [g.strip() for g in genres_str.split(',')]
        all_genres.extend(genres)

genre_counts = Counter(all_genres)
genre_df = pd.DataFrame(genre_counts.most_common(), columns=['Genre', 'Count'])

print("Genre Frequency Distribution:\n")
print(genre_df)

# Movies per genre combination
genre_combo_counts = df_raw['genres'].value_counts().head(15)

print("\nTop 15 Genre Combinations:\n")
for combo, count in genre_combo_counts.items():
    print(f"{combo:<50} {count:>6,}")
```

```python
# Visualize genre distribution (SAVE)
fig = go.Figure([
    go.Bar(
        x=genre_df['Genre'],
        y=genre_df['Count'],
        marker_color=GOLD,
        marker_line_color=BLACK,
        marker_line_width=1
    )
])

fig.update_layout(
    template=plotly_template,
    title='Genre Frequency Distribution',
    xaxis_title='Genre',
    yaxis_title='Count',
    xaxis_tickangle=-45,
    height=500,
    showlegend=False
)

fig.write_html(CHART_DIR / '03_genre_distribution.html')
fig.show()
print(f"✓ Saved: {CHART_DIR / '03_genre_distribution.html'}")
```

## 6. Numerical Features Analysis

### 6.1 Vote Average & Vote Count

```python
# Vote statistics
print("Vote Average Statistics:")
print(df_raw['vote_average'].describe())

print("\nVote Count Statistics:")
print(df_raw['vote_count'].describe())

# Check low vote counts
low_votes = (df_raw['vote_count'] < 50).sum()
print(f"\n⚠ Movies with < 50 votes: {low_votes:,} ({100*low_votes/len(df_raw):.2f}%)")
print("→ These will be filtered out in preprocessing")
```

### 6.2 Runtime Analysis

```python
# Runtime distribution
runtime_clean = df_raw['runtime'].dropna()
runtime_clean = runtime_clean[runtime_clean > 0]  # Remove zeros

print("\nRuntime Statistics:")
print(runtime_clean.describe())

# Outliers
print(f"\nMovies with runtime > 300 min: {(runtime_clean > 300).sum()}")
print(f"Movies with runtime = 0: {(df_raw['runtime'] == 0).sum()}")
```

## 7. Temporal Analysis

```python
# Convert release_date to datetime and extract year
df_raw['release_date_parsed'] = pd.to_datetime(df_raw['release_date'], errors='coerce')
df_raw['release_year'] = df_raw['release_date_parsed'].dt.year

# Movies per year
movies_per_year = df_raw['release_year'].value_counts().sort_index()

# Filter to reasonable range (1900-2024)
movies_per_year = movies_per_year[(movies_per_year.index >= 1900) & (movies_per_year.index <= 2024)]

print(f"\nYear range: {movies_per_year.index.min()}-{movies_per_year.index.max()}")
print(f"Peak year: {movies_per_year.idxmax()} ({movies_per_year.max():,} movies)")
```

```python
# Movies over time (SAVE)
fig = go.Figure([
    go.Scatter(
        x=movies_per_year.index,
        y=movies_per_year.values,
        mode='lines',
        line=dict(color=GOLD, width=2),
        fill='tozeroy',
        fillcolor=f'rgba(234, 179, 8, 0.3)'
    )
])

fig.update_layout(
    template=plotly_template,
    title='Number of Movies Released Per Year (1900-2024)',
    xaxis_title='Year',
    yaxis_title='Number of Movies',
    height=500,
    showlegend=False
)

fig.write_html(CHART_DIR / '06_movies_over_time.html')
fig.show()
print(f"✓ Saved: {CHART_DIR / '06_movies_over_time.html'}")
```

```python
# Average rating over time (SAVE)
avg_rating_per_year = df_raw.groupby('release_year')['vote_average'].mean()
avg_rating_per_year = avg_rating_per_year[(avg_rating_per_year.index >= 1950) & (avg_rating_per_year.index <= 2024)]

fig = go.Figure([
    go.Scatter(
        x=avg_rating_per_year.index,
        y=avg_rating_per_year.values,
        mode='lines',
        line=dict(color=GOLD, width=2)
    )
])

fig.update_layout(
    template=plotly_template,
    title='Average Movie Rating Over Time (1950-2024)',
    xaxis_title='Year',
    yaxis_title='Average Vote Average',
    height=500,
    showlegend=False
)

fig.write_html(CHART_DIR / '07_rating_over_time.html')
fig.show()
print(f"✓ Saved: {CHART_DIR / '07_rating_over_time.html'}")
```

## 9. Data Quality Issues Summary

```python
print("="*80)
print("DATA QUALITY ISSUES SUMMARY")
print("="*80)

issues = []

# 1. Non-released movies
non_released = len(df_raw[df_raw['status'] != 'Released'])
issues.append({
    'Issue': 'Non-Released Movies',
    'Count': non_released,
    'Percentage': f"{100*non_released/len(df_raw):.2f}%",
    'Action': 'Filter out (keep Released only)'
})

# 2. Adult content
adult_movies = len(df_raw[df_raw['adult'] == True])
issues.append({
    'Issue': 'Adult Content',
    'Count': adult_movies,
    'Percentage': f"{100*adult_movies/len(df_raw):.2f}%",
    'Action': 'Filter out'
})

# 3. Low vote counts
low_votes = len(df_raw[df_raw['vote_count'] < 50])
issues.append({
    'Issue': 'Vote Count < 50',
    'Count': low_votes,
    'Percentage': f"{100*low_votes/len(df_raw):.2f}%",
    'Action': 'Filter out (unreliable ratings)'
})

# 4. Missing genres
missing_genres = df_raw['genres'].isna().sum()
issues.append({
    'Issue': 'Missing Genres',
    'Count': missing_genres,
    'Percentage': f"{100*missing_genres/len(df_raw):.2f}%",
    'Action': 'Drop (cannot cluster without genre)'
})

# 5. Runtime outliers
runtime_outliers = len(df_raw[(df_raw['runtime'] == 0) | (df_raw['runtime'] > 300)])
issues.append({
    'Issue': 'Runtime Outliers (0 or >300 min)',
    'Count': runtime_outliers,
    'Percentage': f"{100*runtime_outliers/len(df_raw):.2f}%",
    'Action': 'Filter out'
})

# 6. Budget/Revenue zeros
budget_zeros = (df_raw['budget'] == 0).sum()
issues.append({
    'Issue': 'Budget = 0 (missing)',
    'Count': budget_zeros,
    'Percentage': f"{100*budget_zeros/len(df_raw):.2f}%",
    'Action': 'Convert to NaN, assess viability'
})

revenue_zeros = (df_raw['revenue'] == 0).sum()
issues.append({
    'Issue': 'Revenue = 0 (missing)',
    'Count': revenue_zeros,
    'Percentage': f"{100*revenue_zeros/len(df_raw):.2f}%",
    'Action': 'Convert to NaN, assess viability'
})

issues_df = pd.DataFrame(issues)
print("\n")
print(issues_df.to_string(index=False))

print("\n" + "="*80)
print("These issues will be addressed in the preprocessing pipeline.")
print("="*80)
```

## 11. Key Findings & Recommendations

### Summary of Findings

#### 1. Dataset Composition
- Total movies: ~1.4M rows
- Released movies: ~96% of dataset
- Adult content: ~10% (will be filtered)
- Primary language: English (~65%), followed by French, Spanish, Italian

#### 2. Data Quality
- Vote count: ~98% of movies have <50 votes (unreliable, will filter)
- Budget/Revenue: ~95% missing/zero (will exclude from features)
- Genres: Well-populated (~56% have genres), critical for clustering
- Keywords: Rich semantic information, top keywords align with common themes
- Runtime: Most movies 80-130 min, some outliers >300 min (documentaries/series)

#### 3. Rating Patterns
- Vote average: Mean ~1.6, heavily skewed by movies with no votes
- Strong correlation: vote_count ↔ popularity
- Weak correlation: vote_average ↔ runtime

#### 4. Temporal Trends
- Production growth: Exponential increase from 1980s onward
- Recent years: Dominated by 2010-2024 releases
- Rating stability: Average ratings relatively stable over time

#### 5. Genre Distribution
- Most common: Documentary, Drama, Comedy, Animation, Music
- Genre combinations: Drama frequently paired with other genres

---

### Preprocessing Recommendations

1. **Filter by status:** Keep only "Released" movies
2. **Remove adult content:** Filter out adult=True
3. **Vote count threshold:** Filter vote_count >= 50 for reliable ratings
4. **Runtime bounds:** Keep 1 ≤ runtime ≤ 300 minutes
5. **Handle missing:**
   - Drop rows with missing genres
   - Fill overview/keywords with empty string
   - Drop budget/revenue (>90% missing)
6. **Remove duplicates:** By title + release_year

**Expected retention:** ~2% of original dataset (~25K-30K quality movies)

---

### Feature Engineering Recommendations

1. **Genres:** Multi-hot encode all genres (19 binary features)
2. **Keywords:** Top 50-100 most frequent (multi-hot encoding)
3. **Vote average:** Min-Max normalize to [0,1]
4. **Runtime:** Min-Max normalize to [0,1]
5. **Release year:** Min-Max normalize to [0,1]
6. **Overview:** Optional TF-IDF (compare with/without)

**Total estimated features:** 70-150 depending on configuration
